```python

## Create a new dataframe called "newpop" with columns org by year: 
## total pop, total death pop, total wellpath pop, total wellpath death pop total, public pop, public death pop
```

```python
year = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
```

```python

totalpop = [tpop09, tpop10, tpop11, tpop12, tpop13, tpop14, tpop15, tpop16, tpop17, tpop18, tpop19]
totalpop
```

```python
totaldeaths = [dpop09, dpop10, dpop11, dpop12, dpop13, dpop14, dpop15, dpop16, dpop17, dpop18, dpop19]
totaldeaths
```

```python
illnessdeaths = [ilpop09, ilpop10, ilpop11, ilpop12, ilpop13, ilpop14, ilpop15, ilpop16, ilpop17, ilpop18, ilpop19]
```

```python
wellpathpop = [wpop09, wpop10, wpop11, wpop12, wpop13, wpop14, wpop15, wpop16, wpop17, wpop18, wpop19]
wellpathpop
```

```python
wellpathdeathpop = [wd09, wd10, wd11, wd12, wd13, wd14, wd15, wd16, wd17, wd18, wd19]
wellpathdeathpop
```

```python
publicpop = [ppop09, ppop10, ppop11, ppop12, ppop13, ppop14, ppop15, ppop16, ppop17, ppop18, ppop19]
publicpop
```

```python
publicdeathpop = [pd09, pd10, pd11, pd12, pd13, pd14, pd15, pd16, pd17, pd18, pd19]
publicdeathpop
```

```python
newpop = pd.DataFrame(totalpop, index =[year], columns = ["population"])
newpop["wellpath population"] = wellpathpop
newpop["public population"] = publicpop
newpop["non wellpath population"] = newpop["population"] - newpop["wellpath population"]
newpop["deaths"] = totaldeaths
newpop["wellpath deaths"] = wellpathdeathpop
newpop["public deaths"] = publicdeathpop
newpop["non wellpath deaths"] = newpop["deaths"] - newpop["wellpath deaths"]
newpop["deathrate"] = (newpop["deaths"]/newpop["population"])*1000
newpop["wellpath deathrate"] = (newpop["wellpath deaths"]/newpop["public population"])*1000
newpop["public deathrate"] = (newpop["public deaths"]/newpop["wellpath population"])*1000
newpop["non wellpath deathrate"] = (newpop["non wellpath deaths"]/newpop["non wellpath population"])*1000
newpop


```
